```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi)
ysin = np.sin(x)
ycos = np.cos(x)

plt.plot(x,ysin)
plt.plot(x,ycos)

plt.xlim(x.min() * 1.5 ,x.max() * 1.5)
plt.xticks([-np.pi,-np.pi/2,0,np.pi / 2,np.pi],[r'$-\pi$',r'$-\pi/2$','$0$',r'$\pi/2$','$\pi$'])

plt.show()
```
