```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.grid(axis='y',linewidth=0.3)
plt.axvline(x=0,ymax=0.9,ymin=0.1,linestyle='--')
plt.axhline(y=0,linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

x = np.linspace(-np.pi,np.pi)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.grid(axis='y',linewidth=0.3)
plt.legend(['',''])
plt.axvline(x=0,ymax=0.9,ymin=0.1,linestyle='--')
plt.axhline(y=0,linestyle='--')


plt.axvspan(xmin=0.5,xmax=2.0,alpha=0.3)
plt.axhspan(ymin=0.5,ymax=1.0,alpha=0.3)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

men_means = np.array([90.5, 89.5, 88.7, 88.5, 85.2, 86.6])
women_means = np.array([92.7, 87.0, 90.5, 85.0, 89.5, 89.8])

ind = np.arange(len(men_means))
width=0.2
plt.bar(ind-0.1,men_means,width=width,label='')
plt.bar(ind+0.2,women_means,width=width,label='')
plt.title('')

plt.xticks(ind,["1", "2", "3", "4", "5", "6"])
plt.ylabel('')

# 
plt.axhline(88.5,linestyle='--',linewidth=1.0,label='')


plt.legend(loc='lower right')
plt.show()
```
