```python
import matplotlib.pyplot as plt
import numpy as np

num = 50
x = np.random.rand(num)
y = np.random.rand(num)

plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

num = 50
x = np.random.rand(num)
y = np.random.rand(num)

plt.scatter(x,y,s=(np.random.rand(num) * 30)**2,alpha=0.5)
plt.show()
```
